```python
from manim import *
import numpy as np

class AlphaCarinaLogo(Scene):
    def construct(self):
        arc = Arc(
            radius=4,
            start_angle=200 * DEGREES,
            angle=180 * DEGREES,
            stroke_width=20,
            color=WHITE
        )

        self.play(Create(arc), run_time=2)

        angles = np.linspace(-140, 0, 8)
        nodes = VGroup()

        for a in angles:
            p = arc.point_from_proportion((a + 140) / 140)
            dot = Dot(p, radius=0.08)
            nodes.add(dot)

        self.play(LaggedStartMap(FadeIn, nodes), run_time=2)

        edges = VGroup(
            Line(nodes[i].get_center(), nodes[i+1].get_center())
            for i in range(len(nodes)-1)
        )

        self.play(Create(edges), run_time=2)
        self.wait(1)
```
